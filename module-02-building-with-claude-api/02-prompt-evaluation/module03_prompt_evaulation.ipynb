```python
# Create an API Client
from anthropic import Anthropic

# Initialize the Client
client = Anthropic()

# Select a Model
model = "claude-sonnet-4-0"

def add_user_message(messages, text):
    user_messages = {"role": "user", "content": text}
    messages.append(user_messages)

def add_assistant_message(messages, text):
    assistant_messages = {"role": "assistant", "content": text}
    messages.append(assistant_messages)

def create_message(messages, system_prompt_msg=None, temperature=1.0, stop_sequences=[]):
    parameters = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences,
    }
    if system_prompt_msg:
        parameters["system"] = system_prompt_msg
    message = client.messages.create(**parameters)
    return message.content[0].text
```

```python
# make a starting list of messages
messages = []
# add a user message to the list of messages
add_user_message(messages, "What is the capital of France?")
# create a message using the list of messages
response = create_message(messages)
# print the response
print(response)

# add the assistant's response to the list of messages
add_assistant_message(messages, response) 

# add another user message to the list of messages
add_user_message(messages, "What are the top 3 famous landmarks in there?")
# create a message using the list of messages
response = create_message(messages)
#print the response
print(response)

# add the assistant's response to the list of messages
add_assistant_message(messages, response)

# add user message to the list of messages
add_user_message(messages, "Who is the current president of the country?")
# create a message using the list of messages
response = create_message(messages)
# print the response
print(response)
# add the assistant's response to the list of messages
add_assistant_message(messages, response)
```

```python
messages = []

while True:
    user_input = input("User: ")
    print(f"User: {user_input}")
    add_user_message(messages, user_input)
    response = create_message(messages)
    print(f"Assistant: {response}")
    add_assistant_message(messages, response)
```
